```python
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
input_image=cv.imread('/content/drive/MyDrive/Face_Recognition/robert.png')
```

```python
resized_image=cv.resize(input_image,(160,160))
```

```python
print(resized_image.shape)
```

**1} Horizontal Shift**

```python
h_shift=40
h_shifted=np.roll(resized_image,h_shift,axis=1)
cv2_imshow(h_shifted)
```

**2} Vertical Shift**

```python
v_shift=-50
v_shifted=np.roll(resized_image,v_shift,axis=0)
cv2_imshow(v_shifted)
```

**3} Brightness**


1.   Higher alpha (>1) increases contrast, making dark areas darker and bright

2.   Higher beta (>0) makes the whole image brighter.

3.   Lower alpha (<1) reduces contrast.

4.   Lower beta (<0) darkens the image.


```python
bright=cv.convertScaleAbs(resized_image,alpha=1.2,beta=60)
cv2_imshow(bright)
```

**4} Darkness**

```python
dark=cv.convertScaleAbs(resized_image,alpha=0.7,beta=-30)
cv2_imshow(dark)
```

**5} Blurring**



1.   resized_image: The input image (a NumPy array)

2.   (7,7): The kernel size (width, height). It must be an odd number (e.g., 3, 5, 7).A larger kernel results in a stronger blur.

3.   The standard deviation (sigmaX) along the X-axis. If set to 0, OpenCV calculates it automatically.



```python
blur=cv.GaussianBlur(resized_image,(7,7),0)
cv2_imshow(blur)
```

**6} Flipping**

```python
flip=cv.flip(resized_image,1)
cv2_imshow(flip)
```
