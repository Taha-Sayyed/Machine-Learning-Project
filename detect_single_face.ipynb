```python
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/taha.png")
```

```python
from mtcnn import MTCNN
import cv2

# Initialize face detector
detector = MTCNN()

# Check if image is loaded properly
if img is None:
    print("Error: Image not found or unable to read.")
else:
    # Perform face detection
    output = detector.detect_faces(img)
    print(output)
```

```python
x,y,width,height= output[0]['box']
```

```python
cv2.rectangle(img,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=3)
```

```python
#AIM:Detecting Facial Landmark
```

```python
left_eyeX,left_eyeY=output[0]['keypoints']['left_eye']
right_eyeX,right_eyeY=output[0]['keypoints']['right_eye']
noseX,noseY=output[0]['keypoints']['nose']
mouth_leftX,mouth_leftY=output[0]['keypoints']['mouth_left']
mouth_rightX,mouth_rightY=output[0]['keypoints']['mouth_right']
```

```python
print(left_eyeX,left_eyeY)
print(right_eyeX,right_eyeY)
print(noseX,noseY)
print(mouth_leftX,mouth_leftY)
print(mouth_rightX,mouth_rightY)
```

```python
cv2.circle(img,center=(left_eyeX,left_eyeY),radius=5,color=(255,0,0),thickness=1)
cv2.circle(img,center=(right_eyeX,right_eyeY),radius=5,color=(255,0,0),thickness=1)
cv2.circle(img,center=(noseX,noseY),radius=5,color=(255,0,0),thickness=1)
cv2.circle(img,center=(mouth_leftX,mouth_leftY),radius=5,color=(255,0,0),thickness=1)
cv2.circle(img,center=(mouth_rightX,mouth_rightY),radius=5,color=(255,0,0),thickness=1)
cv2.rectangle(img,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=3)
```

```python
#AIM: To detect landmark on multiple faces
```

```python
img_1=cv2.imread("/content/drive/MyDrive/Colab Notebooks/khan.jpeg")
```

```python
output_1=detector.detect_faces(img_1)
```

```python
output_1
```

```python
for i in output_1:
  x,y,width,height= i['box']
  left_eyeX,left_eyeY=i['keypoints']['left_eye']
  right_eyeX,right_eyeY=i['keypoints']['right_eye']
  noseX,noseY=i['keypoints']['nose']
  mouth_leftX,mouth_leftY=i['keypoints']['mouth_left']
  mouth_rightX,mouth_rightY=i['keypoints']['mouth_right']
  cv2.circle(img_1,center=(left_eyeX,left_eyeY),radius=5,color=(255,0,0),thickness=1)
  cv2.circle(img_1,center=(right_eyeX,right_eyeY),radius=5,color=(255,0,0),thickness=1)
  cv2.circle(img_1,center=(noseX,noseY),radius=5,color=(255,0,0),thickness=1)
  cv2.circle(img_1,center=(mouth_leftX,mouth_leftY),radius=5,color=(255,0,0),thickness=1)
  cv2.circle(img_1,center=(mouth_rightX,mouth_rightY),radius=5,color=(255,0,0),thickness=1)
  cv2.rectangle(img_1,pt1=(x,y),pt2=(x+width,y+height),color=(255,0,0),thickness=3)

```
